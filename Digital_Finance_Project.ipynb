```python
import numpy as np
import matplotlib.pyplot as plt

# Define portfolio weights (from 0 to 1)
weights = np.linspace(0, 1, 100)

# Expected returns (already computed)
E_R1 = df_CAP["Daily Return"].mean() * 252  # Annualized return
E_R2 = df_BNPA["Daily Return"].mean() * 252
E_Rp = df_portfolio_2stocks["Daily Return"].mean() * 252

# Standard deviations (risk)
std_1 = df_CAP["Daily Return"].std() * np.sqrt(252)
std_2 = df_BNPA["Daily Return"].std() * np.sqrt(252)
std_p = df_portfolio_2stocks["Daily Return"].std() * np.sqrt(252)

# Correlation between stocks
correlation = df_CAP["Daily Return"].corr(df_BNPA["Daily Return"])

# Compute portfolio return and risk for different weights
portfolio_returns = []
portfolio_risks = []

for w in weights:
    E_Rpp = w * E_R1 + (1 - w) * E_R2  # Expected return
    std_pp = np.sqrt((w**2 * std_1**2) + ((1 - w)**2 * std_2**2) + (2 * w * (1 - w) * std_1 * std_2 * correlation))  # Risk
    portfolio_returns.append(E_Rpp)
    portfolio_risks.append(std_pp)

# Plot Efficient Frontier
plt.figure(figsize=(10, 6))
plt.plot(portfolio_risks, portfolio_returns, label="Efficient Frontier", color='blue')
plt.scatter(std_1, E_R1, c='red', marker='o', label="Portfolio 1: Stock 1 (CAP.PA)")
plt.scatter(std_2, E_R2, c='green', marker='o', label="Portfolio 2: Stock 2 (BN.PA)")
plt.scatter(std_p, E_Rp, c='purple', marker='o', label="Portfolio 3: Equal weight")

# Labels and title
plt.xlabel("Standard Deviation (Risk)")
plt.ylabel("Expected Return")
plt.title("Efficient Frontier of the Portfolios")
plt.legend()
plt.grid(True)
plt.show()

print(f"E_R1: {E_R1}")
print(f"E_R2: {E_R2}")
print(f"E_Rp: {E_Rp}")
print(f"std_1: {std_1}")
print(f"std_2: {std_2}")
print(f"std_p: {std_p}")
```

# **Question 2.10**

For the Sharpe Ratio we use the formula:

$
\frac{\mathbb{E}[R_a-R_f]}{std\_dev[R_a-R_f]}
$

For the Sortino Ratio we used the formula:

$
\frac{\mathbb{E}[R_a-R_f]}{std\_dev[R_a-R_f]}
$

And for the Sterling Ratio we used the formula showed in class:

$
\frac{Return}{(Avg Max Drawdown) + Benchmark Volatility}
$

Where the benchmark market will be the CAC 40

```python
# Adapting interest rate of the US treasury for a daily interest (13 weeks -> 1 day)

df_IRX['Daily rate']=(((df_IRX['Adj Close']/100)+1)**(1/65)) -1

# Substraction of the Daily Return with the US treasury interest rate (excess return)

df_portfolio_2stocks['Excess Return'] = df_portfolio_2stocks["Daily Return"].sub(df_IRX['Daily rate'])
#, fill_value=0

# Expected value and Standard deviation


E_difference=df_portfolio_2stocks['Excess Return'].mean()

std_difference=df_portfolio_2stocks['Excess Return'].std()

#Sharpe ratio

Sh_ratio=E_difference/std_difference

print(f"The Sharpe ratio of the portfolio is: {Sh_ratio}")


# Sortino ratio

# Convert returns to excess returns over the risk-free rate

excess_return = df_portfolio_2stocks['Excess Return']

# Calculate downside deviation (only negative returns)

downside_return = excess_return[excess_return < 0]

downside_deviation = np.std(downside_return, ddof=1)

# Calculate Sortino Ratio

Sorti_ratio= excess_return.mean() / downside_deviation

print(f"The Sortino ratio of the portfolio is: {Sorti_ratio} ")

# Sterling ratio

# Calculate Maximum Drawdown
def max_drawdown(returns):
    returns=pd.Series(returns)
    cumulative_returns = (1 + returns).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (cumulative_returns - peak) / peak
    return drawdown.min()  # Worst drawdown

# Compute Maximum Drawdowns for rolling periods
rolling_drawdowns = df_portfolio_2stocks["Daily Return"].rolling(window=65).apply(max_drawdown, raw=True).dropna()

# Compute Average of Maximum Drawdowns
avg_max_drawdown = np.abs(rolling_drawdowns.mean())

# Compute Benchmark Volatility (Standard Deviation)
std_CAC40=df_CAC40['Return'].std()

# Calculate Sterling ratio
denominator=avg_max_drawdown+std_CAC40

Sterl_ratio=df_portfolio_2stocks["Daily Return"].mean()/denominator

print(f"The Sterling ratio of the portfolio is: {Sterl_ratio}")
```

The ratios are different because they represent different properties of the risk:
- The Sharpe ratio express the expected value of the return in relation with a general volatility of the return.
- The Sortino ratio on the other hand, only uses the negative part of the excess return, which is what investors are usually interested in.
- Finally, the Sterling ratio uses the "maximum risk of loss" in a certain period, wich is very useful to understand what risk investors need to beware of.

```python
import scipy.stats as stats

# 1. Let's compute the 5% empirical VaR:
var_5_empirical = np.percentile(df_portfolio_2stocks["Daily Return"].dropna(), 5)

# 2. Compute the 1st percentile of the normal distribution
mu = df_portfolio_2stocks["Daily Return"].mean()  # Empirical mean
sigma = df_portfolio_2stocks["Daily Return"].std()  # Empirical standard deviation
var_1_normal = stats.norm.ppf(0.01, loc=mu, scale=sigma)  # 1st percentile of normal distribution

# Print results
print(f"Empirical VaR (5%): {var_5_empirical:.4f}")
print(f"Normal VaR (1%): {var_1_normal:.4f}")
```

Normal VaR (1%):
The 1st percentile of the normal distribution corresponds to a large negative deviation from the mean.
-Interpretation: VaR = -0.0228 means there is a 1% probability of the portfolio losing at least 2.28% in a single day.

Empirical VaR (5%): -0.0154 means:
-There is a 5% chance that the portfolio will lose at least 1.54% in a single day.
-In other words, on the worst 5% of trading days, the portfolio's return is expected to be -1.54% or worse.

```python
# Step 1: Generate 1000 bootstrapped samples from historical daily returns
bootstrapped_returns = np.random.choice(df_portfolio_2stocks["Daily Return"].dropna(), size=1000, replace=True)

# Step 2: Compute the 5% VaR from the simulated dataset
var_5_bootstrap = np.percentile(bootstrapped_returns, 5)

# Print results
print(f"Bootstrapped VaR (5%): {var_5_bootstrap:.4f}")
print(f"Empirical VaR (5%): {var_5_empirical:.4f}")
print(f"Normal VaR (1%): {var_1_normal:.4f}")
```

**Bootstrapped VaR < Empirical VaR**    
Indicates that resampling captured more extreme losses, meaning the true downside risk might be underestimated by historical data alone.
Possible reason: Some extreme losses may not have occurred yet in the historical period but appear when resampling.    

**Bootstrapped VaR > Normal VaR (1%)**    
The normal distribution still predicts even greater downside risk.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load stock data
bnpa = pd.read_csv("BN.PA.csv", index_col="Date", parse_dates=True)
cappa = pd.read_csv("CAP.PA.csv", index_col="Date", parse_dates=True)

# BN.PA management
stock_data = bnpa[['Close']].copy()
stock_data['Return'] = stock_data['Close'].pct_change()

# Create lagged features
for lag in [1, 3, 5, 10]:
    stock_data[f'Return_Lag{lag}'] = stock_data['Return'].shift(lag)

# Compute moving averages
stock_data['MA10'] = stock_data['Close'].rolling(window=10).mean()
stock_data['MA20'] = stock_data['Close'].rolling(window=20).mean()

# Define trading signal (1 = Buy, -1 = Sell, [l,u] = Do Nothing)
l, u = -0.001, 0.001  # Defining the interval bounds
stock_data['Signal'] = np.where(
    stock_data['Return'].shift(-1) > u, 1,
    np.where(stock_data['Return'].shift(-1) < l, -1, 0)
)
# Drop NaN values from shifting
stock_data.dropna(inplace=True)

# Prepare features and labels
X = stock_data[['Return_Lag1', 'Return_Lag3', 'Return_Lag5', 'MA10', 'MA20']]
y = stock_data['Signal']

# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Train Decision Tree Model
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict signals
stock_data['Predicted_Signal'] = model.predict(X)

# Backtest strategy
stock_data['Strategy_Return'] = stock_data['Predicted_Signal'] * stock_data['Return']
cumulative_strategy = (1 + stock_data['Strategy_Return']).cumprod()
cumulative_market = (1 + stock_data['Return']).cumprod()

# Evaluate performance
accuracy = accuracy_score(y_test, model.predict(X_test))
print(f"Model Accuracy: {accuracy:.2f}")

# Plot performance
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(cumulative_strategy, label="Strategy Performance", color='blue')
plt.plot(cumulative_market, label="Market Performance", color='red', linestyle='dashed')
plt.legend()
plt.title("Trading Strategy vs. Market Performance")
plt.show()

```
